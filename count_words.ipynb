```
import re
```

```
def count_words(text):
  # To count how many times each unique word occurs in a text
  counts = dict() # dictionary containing word-count pair

  # To convert text to lower case
  text = text.lower()

  # To split text into tokens(words), leaving out punctuations
  words = re.split(r'[^\w]', text)  # using regex, refer to the expression.
  print('\n'+ 'words: '+ str(words)+'\n')

  # To Count unique words using count dict.
  for word in words:
    if word!= '':
      if word not in counts: 
        counts[word] = 1
      else:
        counts[word] += 1

  return counts
```

```
# Function to open, read the text and use count_words(), sort the 'counts dict'
# This function returns most counted word and its frequency.
def run_count_words():
  f = open("/content/drive/My Drive/my_nlp/sample_text.txt", "r") 
  text = f.read()
  counts = count_words(text)
  print('\n'+'counts dict: '+str(counts)+'\n')
  # Note counts.items() converts a dict into a list of tuples. Also, note the 
  # lambda expression. x[1] is for 2nd element of tuple 
  # reverse=True is for sorting in descending order
  sorted_counts = sorted(counts.items(), key=lambda x:x[1], reverse=True ) 
  print('\n'+'sorted_counts: '+str(sorted_counts)+'\n')
  return sorted_counts[0]



```
